```python
from typing import List
import numpy as np
import scipy.special as sp_spec
import numpy.random as np_rand
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import seaborn as sns
```

# Bernoulli Mixture Model - CAVI and SVI
See the solution pdf for derivation of CAVI updates, ELBO and for notation used throughout the notebook. See theory part of Exercise problems for algorithm psuedo code and more. Updates are written using for-loops to more clearly show relation to derivations, however, you are recommended to use matrix operations.


```python
def generate_data_and_priors(N: int, D: int, K: int, theta_a: float, theta_b: float, pi_alpha: List[int]):
  """
  Generates datapoints and hidden variables given parameters for the priors.
  :param N: Number of datapoints
  :param D: Number of Bernoulli outcomes per datapoint.
  :param K: Number of mixture components.
  :param theta_a: "a" parameter of the Beta(a,b) prior on theta.
  :param theta_b: "b" parameter of the Beta(a,b) prior on theta.
  :param pi_alpha: parameter of the Dirichlet(alpha) prior on pi.
  :return:
  """
  theta = np_rand.beta(theta_a, theta_b, size=K)
  pi = np_rand.dirichlet(pi_alpha)
  x = np.zeros((N,D))
  z = np.zeros(N) # one-hot encoding
  for n in range(N):
    z_n = np_rand.multinomial(1, pi).argmax() # for each datapoint n, sample the component assignment/class variable
    x_n = np.zeros(D)
    for d in range(D):  # sample d times from a bernoulli (=binomial(1,theta)) using the component specific parameter theta_k
      x_n[d] = np_rand.binomial(1, theta[z_n])
    z[n] = z_n
    x[n, :] = x_n

  return x, z, theta, pi

def generate_data(N: int, D: int, K: int, theta: np.ndarray , pi: np.ndarray):
  """
   Generates datapoints given values of the hidden variables. Can be used for more "control" of experiments.
  :param N: Number of datapoints
  :param D: Number of Bernoulli outcomes per datapoint.
  :param K: Number of mixture components.
  :param theta: Values of theta.
  :param pi: Value of pi.
  :return:
  """
  x = np.zeros((N,D))
  z = np.zeros(N) # one-hot encoding
  for n in range(N):
    z_n = np_rand.multinomial(1, pi).argmax() # for each datapoint n, sample the component assignment/class variable
    x_n = np.zeros(D)
    for d in range(D):  # sample d times from a bernoulli (=binomial(1,theta)) using the component specific parameter theta_k
      x_n[d] = np_rand.binomial(1, theta[z_n])
    z[n] = z_n
    x[n, :] = x_n

  return x, z, theta, pi

# Test the functions by running for some simple cases and verify that data and variables look as expected.
N = 100
D = 100
K = 2
theta_a = 1
theta_b = 1
pi_alpha = [1, 1]
theta = np.array([0.1, 0.8])
pi = np.array([0.4, 0.6])
x, z, theta, pi = generate_data(N, D, K, theta, pi)
print(f"First 3 datapoints: {x[0:3, 0:5]}")
print(f"First 3 z: {z[0:3]}")
print(f"theta: {theta}")
print(f"pi: {pi}")
```

## Initialization and ELBO calculation

Initialization strategy and ELBO calculation are the save for CAVI and SVI algorithms so we define them first.

```python
def initialize_q(x, N, K):
  #r_q_init = np.random.random((N,K))
  kmeans =  KMeans(n_clusters=K, random_state=0, n_init=3).fit(x)
  labels = kmeans.labels_
  r_q_init = np.eye(K)[labels] + 2/K  # KMeans init + some uniform noise
  r_q_init = r_q_init / np.sum(r_q_init, axis=1, keepdims=True)  # normalize
  a_q_init = np.linspace(1, K, K)  # initialize coin biases to be evenly spaced (E[theta_k] = a_k/(a_k + b_k)
  b_q_init = np.ones(K)
  alpha_q_init = np.ones(K)
  return r_q_init, a_q_init, b_q_init, alpha_q_init

def calculate_elbo(x, r_q, a_q, b_q, alpha_q):
  x_D = np.sum(x, axis=1)
  N, D = x.shape
  E_log_theta = sp_spec.digamma(a_q) - sp_spec.digamma(a_q + b_q)
  E_log_1_minus_theta = sp_spec.digamma(b_q) - sp_spec.digamma(a_q + b_q)
  E_log_pi = sp_spec.digamma(alpha_q) - sp_spec.digamma(np.sum(alpha_q))
  alpha_0 = np.sum(alpha_q)
  CR_likelihood = np.sum(r_q * (np.einsum('n,k->nk', x_D, E_log_theta) + np.einsum('n,k->nk', D - x_D, E_log_1_minus_theta)))
  CR_Z = np.sum(r_q * E_log_pi)
  eps = 0.0001  # for numerical stability to avoid -inf
  CR_pi = -np.log(multivariate_beta_function(alpha_q) + eps) + np.sum((alpha_q - 1) * E_log_pi)
  CR_theta = np.sum((a_q-1) * E_log_theta + (b_q-1) * E_log_1_minus_theta - sp_spec.betaln(a_q, b_q))
  H_Z = np.sum(r_q * np.log(r_q + eps))
  H_pi = np.log(multivariate_beta_function(alpha_q) + eps) + (alpha_0 - K) * sp_spec.digamma(alpha_0) - \
         np.sum((alpha_q - 1) * sp_spec.digamma(alpha_q))
  ln_Beta_ab = sp_spec.betaln(a_q, b_q) + eps  # Important to use scipy.special.betaln instead of np.log(scipy.special.beta)
  H_theta = np.sum(ln_Beta_ab - (a_q - 1) * sp_spec.digamma(a_q) -
                   (b_q - 1) * sp_spec.digamma(b_q) + (a_q + b_q + 2) * sp_spec.digamma(a_q + b_q))
  elbo = CR_likelihood + CR_Z + CR_pi + CR_theta + H_Z + H_pi + H_theta
  return elbo

N = 10
K = 5
x = np.random.random((N, K))
r_q_init, a_q_init, b_q_init, alpha_q_init = initialize_q(x, N, K)
```

### Helper functions

```python
def multivariate_beta_function(a):
    eps = 0.0001  # for numerical stability to avoid -inf
    return np.exp(np.sum(sp_spec.gammaln(a + eps) - sp_spec.gammaln(np.sum(a + eps))))
```

## CAVI update equations and CAVI algorithm

```python
def update_q_Z(x, a_q, b_q, alpha_q):
  """
  Implements the CAVI update equation of q(Z) derived in the solution of the exercise.
  """
  N, D = x.shape
  E_log_theta = sp_spec.digamma(a_q) - sp_spec.digamma(a_q + b_q)
  E_log_1_minus_theta = sp_spec.digamma(b_q) - sp_spec.digamma(a_q + b_q)
  E_log_pi = sp_spec.digamma(alpha_q) - sp_spec.digamma(np.sum(alpha_q))
  log_rho = np.zeros((N, K))
  for n in range(N):  # Can be written using matrix operations instead
    log_rho_n = np.sum(x[n]) * E_log_theta + (D - np.sum(x[n])) * E_log_1_minus_theta + E_log_pi
    log_rho[n] = log_rho_n

  r_q = np.exp(log_rho - sp_spec.logsumexp(log_rho, axis=1, keepdims=True))
  return r_q

def update_q_pi(r_q, alpha_prior):
  E_Z = r_q
  alpha_q = np.sum(E_Z, axis=0) + alpha_prior  # size K array
  return alpha_q


def update_q_theta(x, r_q, a_prior, b_prior):
  E_Z = r_q
  a_q = np.zeros(K)
  b_q = np.zeros(K)
  x_D = np.sum(x, axis=1)  # Can be calculated outside for-loop of CAVI alg. for effeciency.
  for k in range(K):
    a_k = np.sum(E_Z[:,k] * x_D) + a_prior
    b_k = np.sum(E_Z[:,k] * (D - x_D)) + b_prior
    a_q[k] = a_k
    b_q[k] = b_k

  return a_q, b_q


def CAVI_algorithm(x, K, n_iter, a_prior, b_prior, alpha_prior):
  N, D = x.shape
  r_q, a_q, b_q, alpha_q = initialize_q(x, N, K)
  # Store output per iteration
  elbo = np.zeros(n_iter)
  r_q_out = np.zeros((n_iter, N, K))
  a_q_out = np.zeros((n_iter, K))
  b_q_out = np.zeros((n_iter, K))
  alpha_q_out = np.zeros((n_iter, K))
  for i in range(0, n_iter):
    # CAVI updates
    # q(Z) update
    r_q = update_q_Z(x, a_q, b_q, alpha_q)

    # q(theta) update
    a_q, b_q = update_q_theta(x, r_q, a_prior, b_prior)

    # q(pi) update
    alpha_q = update_q_pi(r_q, alpha_prior)

    # ELBO
    elbo[i] = calculate_elbo(x, r_q, a_q, b_q, alpha_q)

    # outputs
    r_q_out[i] = r_q
    a_q_out[i] = a_q
    b_q_out[i] = b_q
    alpha_q_out[i] = alpha_q

  return r_q_out, a_q_out, b_q_out, alpha_q_out, elbo

# Run optimization
"""
Run and tweak the parameters and examine the outcomes. How does N effect the variational posteriors/distributions? How does the priors effect the variational posteriors? What happens if the theta_k:s are closer together?
"""

N = 100
D = 100
K = 3
n_iter = 20
a_prior = 1.
b_prior = 1.
alpha_prior = np.ones(K) * 1.0
theta = np.array([0.1, 0.8, 0.5])
pi = np.array([0.3, 0.2, 0.5])
x, z, theta, pi = generate_data(N, D, K, theta, pi)
r_q_out, a_q_out, b_q_out, alpha_q_out, elbo = CAVI_algorithm(x, K, n_iter, a_prior, b_prior, alpha_prior)
np.printoptions(precision=2)
print(f"Print results (check for label switching).")
print(f"z :{z[0:10]}")
# As q(Z) is our variational posterior and is a Categorical, argmax corresponds to our MAP estimates
MAP_assignments = r_q_out[-1].argmax(axis=1)
print(f"r_q :{MAP_assignments[0:10]}")
print(f"ARI :{adjusted_rand_score(z, MAP_assignments)}")   # clustering metric that is independent of label switching
print(f"Expected theta: {a_q_out[-1] / (a_q_out[-1] + b_q_out[-1])}")  # E_q[theta] = a / (a + b) from expected value of Beta r.v.
print(f"True theta: {theta}")
print(f"Expected pi: {alpha_q_out[-1] / np.sum(alpha_q_out[-1])}")
print(f"True pi: {pi}")
```

## SVI tailored updates and SVI algorithm

```python
def update_q_Z_batch(batch, x, a_q, b_q, alpha_q):
  """
  Implements the update equation of q(Z) derived in the solution of the exercise.
  Updates only on the datapoint(s) in batch.
  """
  N, D = x.shape
  K = a_q.shape[0]
  S = batch.shape[0]
  E_log_theta = sp_spec.digamma(a_q) - sp_spec.digamma(a_q + b_q)
  E_log_1_minus_theta = sp_spec.digamma(b_q) - sp_spec.digamma(a_q + b_q)
  E_log_pi = sp_spec.digamma(alpha_q) - sp_spec.digamma(np.sum(alpha_q))
  log_rho = np.zeros((S, K))
  for i, n in enumerate(batch):  # Can be written using matrix operations instead
    x_n_D = np.sum(x[n])  # written as x_n^D in the exercise solution notes
    log_rho_n = x_n_D * E_log_theta + (D - x_n_D) * E_log_1_minus_theta + E_log_pi
    log_rho[i] = log_rho_n

  r_q_batch = np.exp(log_rho - sp_spec.logsumexp(log_rho, axis=1, keepdims=True))
  return r_q_batch

def update_q_pi_batch(batch, r_q, alpha_prior):
  """
  Implements the update equation of q(pi) derived in the solution of the exercise.
  """
  E_Z_batch = r_q[batch]
  N, K = r_q.shape
  alpha_q = E_Z_batch * N + alpha_prior  # size K array
  return alpha_q

def update_q_theta_batch(batch, x, r_q, a_prior, b_prior):
  """
  Implements the update equation of q(theta_k) derived in the solution of the exercise.
  NOTE: opposed to the updates in CAVI algorithm in the cell above, the loop over k bellow does
  """
  E_Z_batch = r_q[batch]
  S = batch.shape[0]
  a_q_batch = np.zeros((S, K))
  b_q_batch = np.zeros((S, K))
  x_D_batch = np.sum(x[batch, :], axis=1)
  for k in range(K):
    a_k = E_Z_batch[:,k] * x_D_batch * N + a_prior
    b_k = E_Z_batch[:,k] * N * (D - x_D_batch) + b_prior
    a_q_batch[:, k] = a_k
    b_q_batch[:, k] = b_k

  return a_q_batch, b_q_batch

def calculate_elbo(x, r_q, a_q, b_q, alpha_q):
  x_D = np.sum(x, axis=1)
  N, D = x.shape
  E_log_theta = sp_spec.digamma(a_q) - sp_spec.digamma(a_q + b_q)
  E_log_1_minus_theta = sp_spec.digamma(b_q) - sp_spec.digamma(a_q + b_q)
  E_log_pi = sp_spec.digamma(alpha_q) - sp_spec.digamma(np.sum(alpha_q))
  alpha_0 = np.sum(alpha_q)
  CE_likelihood = np.sum(r_q * (np.einsum('n,k->nk', x_D, E_log_theta) + np.einsum('n,k->nk', D - x_D, E_log_1_minus_theta)))
  CE_Z = np.sum(r_q * E_log_pi)
  eps = 0.0001  # for numerical stability to avoid -inf
  CE_pi = -np.log(multivariate_beta_function(alpha_q) + eps) + np.sum((alpha_q - 1) * E_log_pi)
  CE_theta = np.sum((a_q-1) * E_log_theta + (b_q-1) * E_log_1_minus_theta - sp_spec.betaln(a_q, b_q))
  H_Z = np.sum(r_q * np.log(r_q + eps))
  H_pi = np.log(multivariate_beta_function(alpha_q) + eps) + (alpha_0 - K) * sp_spec.digamma(alpha_0) - \
         np.sum((alpha_q - 1) * sp_spec.digamma(alpha_q))
  ln_Beta_ab = sp_spec.betaln(a_q, b_q) + eps  # Important to use scipy.special.betaln instead of np.log(scipy.special.beta)
  H_theta = np.sum(ln_Beta_ab - (a_q - 1) * sp_spec.digamma(a_q) -
                   (b_q - 1) * sp_spec.digamma(b_q) + (a_q + b_q + 2) * sp_spec.digamma(a_q + b_q))
  elbo = CE_likelihood + CE_Z + CE_pi + CE_theta + H_Z + H_pi + H_theta
  return elbo

def SVI_algorithm(x, K, n_iter, S, a_prior, b_prior, alpha_prior):
  N, D = x.shape
  r_q, a_q, b_q, alpha_q = initialize_q(x, N, K)
  # Store output per iteration
  elbo = np.zeros(n_iter)
  r_q_out = np.zeros((n_iter, N, K))
  a_q_out = np.zeros((n_iter, K))
  b_q_out = np.zeros((n_iter, K))
  alpha_q_out = np.zeros((n_iter, K))

  # step size, rho, related function
  delay = 1 #int(n_iter/10)
  forgetting_rate = 0.6
  rho = lambda t: (t + delay)**(-forgetting_rate)
  for t in range(0, n_iter):
    # sample datapoint and set step size, rho. We are only considering batch size, S = 1 here
    batch = np.random.randint(0, N, size=S)
    print(batch)
    rho_t = rho(t)

    ###### SVI algorithm updates ######
    # Update locals on batch
    # q(Z) update
    r_q_batch = update_q_Z_batch(batch, x, a_q, b_q, alpha_q)
    r_q[batch] = r_q_batch

    # Update intermediate global parameters and take a Robbins-Monro step

    # q(theta) update
    a_q_batch, b_q_batch = update_q_theta_batch(batch, x, r_q, a_prior, b_prior)
    a_q = (1 - rho_t) * a_q + rho_t * 1/S * np.sum(a_q_batch, axis=0)  # Note this is simplifies to step 7. in fig 4. in Hoffman when S=1
    b_q = (1 - rho_t) * b_q + rho_t * 1/S * np.sum(b_q_batch, axis=0)

    # q(pi) update
    alpha_q_batch = update_q_pi_batch(batch, r_q, alpha_prior)
    alpha_q = (1 - rho_t) * alpha_q + rho_t * 1/S * np.sum(alpha_q_batch, axis=0)

    # ELBO
    elbo[t] = calculate_elbo(x, r_q, a_q, b_q, alpha_q)

    # outputs
    r_q_out[t] = r_q
    a_q_out[t] = a_q
    b_q_out[t] = b_q
    alpha_q_out[t] = alpha_q

  return r_q_out, a_q_out, b_q_out, alpha_q_out, elbo

# Run optimization
"""
Run and tweak the parameters and examine the outcomes. How does N effect the variational posteriors/distributions?
How does the priors effect the variational posteriors? What happens if the theta_k:s are closer together?
"""

N = 1000
D = 100
K = 3
S = 50  # batch size <--- need batch size >> 1 to get good results
n_iter = 1000
a_prior = 1.
b_prior = 1.
alpha_prior = np.ones(K) * 1.0
theta = np.array([0.1, 0.8, 0.5])
pi = np.array([0.3, 0.2, 0.5])
x, z, theta, pi = generate_data(N, D, K, theta, pi)
r_q_out, a_q_out, b_q_out, alpha_q_out, elbo = SVI_algorithm(x, K, n_iter, S, a_prior, b_prior, alpha_prior)
np.printoptions(precision=2)
print(f"Print results (check for label switching).")
print(f"z :{z[0:10]}")
# As q(Z) is our variational posterior and is a Categorical, argmax corresponds to our MAP estimates
MAP_assignments = r_q_out[-1].argmax(axis=1)
print(f"q(Z) MAP: {MAP_assignments[0:10]}")
print(f"ARI: {adjusted_rand_score(z, MAP_assignments)}")   # clustering metric that is independent of label switching
print(f"E_q[theta]: {a_q_out[-1] / (a_q_out[-1] + b_q_out[-1])}")  # E_q[theta] = a / (a + b) from expected value of Beta r.v.
print(f"True theta: {theta}")
print(f"E_q[pi]: {alpha_q_out[-1] / np.sum(alpha_q_out[-1])}")
print(f"True pi: {pi}")
```

## SVI vs CAVI comparisons
Here we run experiments on simulated data using the CAVI algorithm and SVI algorithm. The algorithms are then compared in terms of runtime, ELBO and relation to true parameters.
Notice, for some configurations of the simulated data, SVI is both faster and yields better accuracy. Particularly, SVI is substantially faster for large N (approx. > 10**4) and large K (approx. K > 5). In practice, the batch size S is required to be larger than 1 to get good convergence and SVI needs to run for more iterations than SVI. The benefits of SVI are more clear for more complicated models, e.g. Topic models, than for this rather simple Beroulli mixture model.

Simulate data and latent variables and parameters from priors

```python
import time
data_seed = 20
np.random.seed(data_seed)
N = 10**4
D = 5 * 10**2
evenly_spread_out_cluster_components = True
K = 7

# Data generation parameters
pi_alpha_sim = np.ones(K) * 3.

if evenly_spread_out_cluster_components: # Use to avoid clusters being too similar for accurate clustering
  theta = np.linspace(0, 1, K)
  pi = np_rand.dirichlet(pi_alpha_sim)
  x, z, theta, pi = generate_data(N, D, K, theta, pi)
else:
  theta_a_sim = 1.
  theta_b_sim = 1.
  x, z, theta, pi = generate_data_and_priors(N, D, K, theta_a_sim, theta_b_sim, pi_alpha_sim)

print("Sanity check on clusters: if theta's are too close clustering will be difficult")
print(f"Sampled theta: {theta}")
print(f"Sampled pi: {pi}")
```

 ### Visualize optimization

Plots the parameters $a_k$ and $b_k$ of each $ q(\theta_k) $ for different iterations and visualizes the distribution of $q(\theta_k)$ for one k and different iterations.

```python
# Find suitable limits for the x and y axis of the parameter plots
xlim_a_min = a_q_out_svi.min()
xlim_a_max = a_q_out_svi.max()
ylim_b_min = b_q_out_svi.min()
ylim_b_max = b_q_out_svi.max()


plt.rcParams['text.usetex'] = False

# Choose iterations to plot and k to visualize distribution of
iterations_to_plot = [1, 30, 50]
k_dist_vis = 1

n_iter_to_plot = len(iterations_to_plot)
fig, axs = plt.subplots(2, n_iter_to_plot, figsize=(10, 6))
import scipy.stats as sp_stats
for i in range(0, n_iter_to_plot):
  j = iterations_to_plot[i]
  axs[0, i].scatter(a_q_out_svi[j], b_q_out_svi[j])
  axs[0, i].set_title(f'$a_k, b_k$ of $q(\\theta_k)$ at iter: {j}')
  axs[0, i].set_xlim(int(xlim_a_min) - xlim_a_min/2, int(xlim_a_max) + xlim_a_max/2)
  axs[0, i].set_ylim(int(ylim_b_min) - ylim_b_min/2, int(ylim_b_max) + ylim_b_max/2)


  q_theta_1 = sp_stats.beta(a_q_out_svi[i, k_dist_vis], b_q_out_svi[i, k_dist_vis])
  q_theta_1_samples = q_theta_1.rvs(size=1000)
  axs[1, i].hist(q_theta_1_samples, density=True, bins='auto')


  if i == 0:
    axs[0, i].set_xlabel('a')
    axs[0, i].set_ylabel('b')
    axs[1, i].set_ylabel('N samples')

plt.show()
```
