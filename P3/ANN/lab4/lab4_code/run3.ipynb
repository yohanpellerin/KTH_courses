```python
from rbm import RestrictedBoltzmannMachine
from util import *
import matplotlib.pyplot as plt

```

```python
image_size = [28,28]
train_imgs,train_lbls,test_imgs,test_lbls = read_mnist(dim=image_size, n_train=60000, n_test=10000)
list_of_neurons = [500,400,300,200]
recon_erro =[]
for i in list_of_neurons:
    rbm = RestrictedBoltzmannMachine(ndim_visible=image_size[0]*image_size[1],
                                     ndim_hidden=i,
                                     is_bottom=True,
                                     image_size=image_size,
                                     is_top=False,
                                     n_labels=10,
                                     batch_size=20
    )
    
    rbm.cd1(visible_trainset=train_imgs, n_iterations=10)
    recon_erro.append(rbm.recon_error)
    print("Number of neurons: ", i)
    print("Reconstruction error: ", rbm.recon_error)
    print("")

plt.plot(list_of_neurons, recon_erro)
plt.xlabel('Number of neurons')
plt.ylabel('Reconstruction Error')
plt.title('Reconstruction Error vs Number of Neurons')
plt.show()

```
